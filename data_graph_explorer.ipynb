```python
url = "/content/student.csv"

student_data = pd.read_csv(url)
```

```python
student_data.head(2)
```

```python
cols = student_data.columns
print(cols)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

student_data['Grade'] = label_encoder.fit_transform(student_data['Grade'])
student_data['Additional_Work'] = label_encoder.fit_transform(student_data['Additional_Work'])
student_data['Sports_activity'] = label_encoder.fit_transform(student_data['Sports_activity'])
student_data['Attendance'] = label_encoder.fit_transform(student_data['Attendance'])
student_data['Reading'] = label_encoder.fit_transform(student_data['Reading'])
student_data['Notes'] = label_encoder.fit_transform(student_data['Notes'])
student_data['Listening_in_Class'] = label_encoder.fit_transform(student_data['Listening_in_Class'])
student_data['Project_work'] = label_encoder.fit_transform(student_data['Project_work'])

student_data.head()
```

```python
student_data.describe()
```

```python
x = student_data['Grade'].to_numpy()
y = student_data['Weekly_Study_Hours'].to_numpy()

xmin = x.min() - 2
xmax = x.max() + 2
ymin = y.min() - 2
ymax = y.max() + 2

fig, ax = plt.subplots()

plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'k')
plt.plot([0,0],[ymin,ymax], 'k')
plt.plot(x, y, 'rx')
plt.xlabel('Grade')
plt.ylabel('Weekly_Study_Hours')
plt.show()
```
